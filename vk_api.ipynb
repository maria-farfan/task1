```python
import vk_api
import pandas as pd
```

```python
login, password = '--', '--'
vk_session = vk_api.VkApi(login, password)
```

```python
try:
    vk_session.auth(token_only=True)
except vk_api.AuthError as error_msg:
    print(error_msg)
```

```python
vk = vk_session.get_api()
```

```python
vk.wall.get(owner_id = -pub_id1, count=1, \
            filter = 'all')['items'][0]
```
